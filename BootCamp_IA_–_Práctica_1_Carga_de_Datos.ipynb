```python
# --- JSON
json_url = "https://jsonplaceholder.typicode.com/users"
df_json = pd.read_json(json_url)
print("JSON example:")
print(df_json[['id', 'name', 'email']].head(), "\n")

```

```python
# --- Excel
excel_url = "https://file-examples.com/wp-content/uploads/2017/02/file_example_XLSX_10.xlsx"
df_excel = pd.read_excel(excel_url)
print("Excel example:")
print(df_excel.head())
```
